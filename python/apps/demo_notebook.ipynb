```python
# Imports!

import pyMaCh3
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Let's generate a MaCh3 Instance

yaml_config = "jointFitDUNE.yaml"

# Running this will take time
mach3_instance = pyMaCh3.MaCh3Instance(yaml_config)
```

```python
# Okay now we've set up we can play around with things!

# To start with let's make simple program to get event rates
from typing import Iterable
def get_event_rates(mach3_obj: pyMaCh3.MaCh3Instance)->Iterable[float]:
    event_hists = mach3_obj.get_event_hists
    # Now we have the event hists we can do some looping!
    event_list = np.zeros(len(event_hists), dtype=float) # make a big old list of zeros
   
    # Now just loop over hist
    for i, hists in enumerate(event_hists):
        bin_content = hists[0]
        bin_edges = hists[1]
        
        #  Gets event rates [integral of our hist]
        event_list[i] = sum([(bin_edges[i+1]-bin_edges[i])*bin_content[i] for i in range(len(bin_content))])
    
    return event_list
```

```python
osc_vals = mach3_instance.get_nominal_values()


# set unoscillated values to 0
un_osc_vals = osc_vals
un_osc_vals[-6:] = [0]*(len(un_osc_vals)-6)
mach3_instance.set_nominal_values(un_osc_vals)

unosc_event_rates = get_event_rates(mach3_instance)


mach3_instance.set_nominal_values(osc_vals)
osc_event_rates=get_event_rates(mach3_instance)


print(f"UnOsc Event rates : {unosc_event_rates}")
print(f"Osc Event rates : {osc_event_rates}")
```
